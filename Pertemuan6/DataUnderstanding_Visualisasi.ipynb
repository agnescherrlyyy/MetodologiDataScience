```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


negara = ('Argentina', 'Brazil', 'Chile', 'Denmark', 'Finlandia', 'German', 'Hongaria', 'Indonesia', 'Jepang', 'Kamerun')
populasi = (45380000, 212600000, 19120000, 5381000, 5531000, 83240000, 975000, 273500000, 12580000, 2655000)

df = pd.DataFrame({'Country': negara, 'Population': populasi,})
df.sort_values(by='Population', inplace = True)

kor_x = np.arange(len(df))
plt.bar(kor_x, df['Population'], tick_label=df['Country'])
plt.xticks(rotation=90)
plt.ylabel('Populasi (Juta)')
plt.title('Populasi 10 Negara di Dunia')
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


negara = ('Argentina', 'Brazil', 'Chile', 'Denmark', 'Finlandia', 'German', 'Hongaria', 'Indonesia', 'Jepang', 'Kamerun')
populasi = (45380000, 212600000, 19120000, 5381000, 5531000, 83240000, 975000, 273500000, 12580000, 2655000)

df = pd.DataFrame({'Country': negara, 'Population': populasi})
df.sort_values(by='Population', ascending= False, inplace= True)

kor_x = np.arange(len(df))
plt.bar(kor_x, df['Population'], tick_label=df['Country'])
plt.xticks(rotation=90)
plt.ylabel('Populasi (Juta)')
plt.title('Populasi 10 Negara di Dunia')
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


negara = ('Argentina', 'Brazil', 'Chile', 'Denmark', 'Finlandia', 'German', 'Hongaria', 'Indonesia', 'Jepang', 'Kamerun')
populasi = (45380000, 212600000, 19120000, 5381000, 5531000, 83240000, 975000, 273500000, 12580000, 2655000)

df = pd.DataFrame({'Country': negara, 'Population': populasi,})
df.sort_values(by='Population', inplace = True)

kor_x = np.arange(len(df))
warna = ['#0000FF' for _ in range(len(df))]
warna[-3] = '#FF0000'


plt.bar(kor_x, df['Population'], tick_label=df['Country'], color=warna)
plt.xticks(rotation=90)
plt.ylabel('Populasi (Juta)')
plt.title('Populasi 10 Negara di Dunia')
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


negara = ('Argentina', 'Brazil', 'Chile', 'Denmark', 'Finlandia', 'German', 'Hongaria', 'Indonesia', 'Jepang', 'Kamerun')
populasi = (45380000, 212600000, 19120000, 5381000, 5531000, 83240000, 975000, 273500000, 12580000, 2655000)

df = pd.DataFrame({'Country': negara, 'Population': populasi,})
df.sort_values(by='Population', inplace = True)

kor_x = np.arange(len(df))
warna = ['#0000FF' for _ in range(len(df))]
warna[-3] = '#FF0000'

#Membesarkan Gambar
plt.figure(figsize=(20,10))

plt.bar(kor_x, df['Population'], tick_label=df['Country'], color=warna)
plt.xticks(rotation=90)
plt.ylabel('Populasi (Juta)')
plt.title('Populasi 10 Negara di Dunia')
plt.show
```
