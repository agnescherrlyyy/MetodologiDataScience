```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


data_makananpenutup = ('Donat', 'Pastel', 'Kue Coklat', 'Ice Cream', 'Puding Vanila', 'Brownies', 'Puding Strawberi', 'Puding Coklat', 'Ice Cream Nutela',
          'Kue Coklat-Keju', 'Kue Wajik', 'Kue Sus', 'Mochi')
jumlah_penjualan = (14, 5, 12, 19, 6, 8, 12, 9, 10, 17, 2, 9, 13 )

df = pd.DataFrame({'Dessert': data_makananpenutup, 'Sale': jumlah_penjualan})
df.sort_values(by='Sale', inplace = True)

kor_x = np.arange(len(df))
warna = ['#008B8B' for _ in range(len(df))]
warna[-11] = '#556B2F'
warna[-12] = '#556B2F'
warna[-13] = '#556B2F'

#Membesarkan Gambar
plt.figure(figsize=(20,10))

plt.bar(kor_x, df['Sale'], tick_label=df['Dessert'], color=warna)
plt.xticks(rotation=90)
plt.ylabel('Jumlah Penjulan')
plt.title('Penutup Makanan yang Banyak diminati')
plt.show
```
